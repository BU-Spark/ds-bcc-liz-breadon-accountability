```python
import pandas as pd
import os

file_paths = [
    '2022-2023 School Year Reports/Bay State College.xlsx',
    '2022-2023 School Year Reports/Berklee College of Music-Boston Conservatory.xlsx',
    '2022-2023 School Year Reports/Boston Architectural College.xlsx',
    '2022-2023 School Year Reports/Boston Baptist College.xlsx',
    '2022-2023 School Year Reports/Boston College WCAS.xlsx',
    '2022-2023 School Year Reports/Boston College.xlsx',
    '2022-2023 School Year Reports/Boston University.xlsx',
    '2022-2023 School Year Reports/Emerson College.xlsx',
    '2022-2023 School Year Reports/Emmanuel College.xlsx',
    '2022-2023 School Year Reports/Fisher College.xlsx',
    '2022-2023 School Year Reports/Franklin Institute of Technology.xlsx',
    '2022-2023 School Year Reports/Harvard University.xlsx',
    '2022-2023 School Year Reports/MA College of Pharmacy and Health Sciences.xlsx',
    '2022-2023 School Year Reports/Massachusetts College of Art and Design.xlsx',
    '2022-2023 School Year Reports/Massachusetts Institute of Technology.xlsx',
    '2022-2023 School Year Reports/MGH Institute of Health Professions.xlsx',
    '2022-2023 School Year Reports/New England College of Optometry.xlsx',
    '2022-2023 School Year Reports/New England Conservatory.xlsx',
    '2022-2023 School Year Reports/New England Law.xlsx',
    '2022-2023 School Year Reports/Northeastern University.xlsx',
    '2022-2023 School Year Reports/Sattler College.xlsx',
    '2022-2023 School Year Reports/SHOWA Boston.xlsx',
    '2022-2023 School Year Reports/Simmons University.xlsx',
    '2022-2023 School Year Reports/St. John\'s Seminary.xlsx',
    '2022-2023 School Year Reports/Suffolk University.xlsx',
    '2022-2023 School Year Reports/Tufts University.xlsx',
    '2022-2023 School Year Reports/University of Massachusetts-Boston.xlsx',
    '2022-2023 School Year Reports/Urban College of Boston.xlsx',
    '2022-2023 School Year Reports/Wentworth Institute of Technology.xlsx'
]

all_data = []

# Define the columns we want to standardize across files
desired_columns = [
    "6a. Street #",
    "6b. Street Name",
    "6c. Street Suffix",
    "6d. Unit #",
    "6e. Zip",
    "7. Undergraduate (U) or Graduate (G)",
    "8. Full-time (FT) or Part-time (PT)",
    "9. At-Home or Not-at-Home",
    "9. 5 or More Undergrads/Unit (Y/N)",  # New column for '5 or More Undergrads/Unit'
    "University",  # New column to indicate which file the data came from
    "Year"  # New column to indicate the year
]

# Loop through each file and process
for file_path in file_paths:
    file_name = os.path.basename(file_path).replace('.xlsx', '')  # Remove the .xlsx extension
    
    # Read 'Student Addresses' sheet (Sheet 2) from the Excel file
    try:
        df = pd.read_excel(file_path, sheet_name='Student Addresses', skiprows=2)
        
        # Rename columns if necessary to match the desired format
        df.columns = [
            "6a. Street #",
            "6b. Street Name",
            "6c. Street Suffix",
            "6d. Unit #",
            "6e. Zip",
            "7. Undergraduate (U) or Graduate (G)",
            "8. Full-time (FT) or Part-time (PT)",
            "9. At-Home or Not-at-Home"
        ]
        
        # Check if the column for '5 or More Undergrads/Unit (Y/N)' exists, add it if missing
        if '9. 5 or More Undergrads/Unit (Y/N)' not in df.columns:
            df['9. 5 or More Undergrads/Unit (Y/N)'] = 'N/A'  # If the column doesn't exist, fill with 'N/A'
        
        # Add a new column to keep track of which file the data came from (without .xlsx)
        df['University'] = file_name
        
        # Add a new column for the year
        df['Year'] = '2022-2023'
        
        # Append the dataframe to the list
        all_data.append(df)
    
    except Exception as e:
        print(f"Error processing file {file_name}: {e}")

# Combine all dataframes into one
combined_df = pd.concat(all_data, ignore_index=True)

# Reorder the columns to match the desired format
combined_df = combined_df[desired_columns]

# Save the combined dataframe to a new Excel file
output_file = 'formatted_combined_student_addresses_2022_2023.xlsx'
combined_df.to_excel(output_file, index=False)

output_file
```
